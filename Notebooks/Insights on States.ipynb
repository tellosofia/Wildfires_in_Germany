```python
def get_df_from_link(url):
  path = 'https://drive.google.com/uc?id='+url.split('/')[-2]
  df = pd.read_csv(path)
  return df
```

```python
air_temp = get_df_from_link('https://drive.google.com/file/d/1mD2rFwVxkpkCXffKkbaw-UIaGe7v6izw/view?usp=sharing')
fire_causes = get_df_from_link('https://drive.google.com/file/d/1A4i83FZrcw9lbxeWd6CivZNPxhCyGD9z/view?usp=sharing')
negligence = get_df_from_link('https://drive.google.com/file/d/1OU5-eubCLh3bH94YvMfeDJd0TSsMCGvl/view?usp=sharing')
precipitation = get_df_from_link('https://drive.google.com/file/d/13V6aCJdhSo4Ob0SE9PPndcffLaNfI5PT/view?usp=sharing')
wildfires = get_df_from_link('https://drive.google.com/file/d/1cwI7o9BgzcxzKBfJhOPRqIhtU9RgTvrA/view?usp=sharing')
```

```python
# first we create copies just in case
air_temp_df = air_temp.copy()
fire_causes_df = fire_causes.copy()
negligence_df = negligence.copy()
precipitation_df = precipitation.copy()
wildfires_df = wildfires.copy()
```

```python
# we need to split the year/month column on the base dataframe first

# first we convert the date column to datetime
wildfires_df['year/month'] = pd.to_datetime(wildfires_df['year/month'])

# then we split the year/month column usind dt function
wildfires_df['year'] = wildfires_df['year/month'].dt.year
wildfires_df['month'] = wildfires_df['year/month'].dt.month

# then we drop the year/month column
wildfires_df = wildfires_df.drop('year/month', axis=1) # axis = 1 for columns
```

```python
# let's rearrange the columns
wildfires_df = wildfires_df[['year', 'month', 'county', 'amount_of fires', 'burnt_area(ha)']]

# and change names
wildfires_df = wildfires_df.rename(columns={'county':'state', 'amount_of fires':'fire qty', 'burnt_area(ha)':'burnt_area (ha)'})

display(wildfires_df.head(1))
```

```python
# let's check for duplicates

display(air_temp.duplicated().sum(), fire_causes.duplicated().sum(), negligence.duplicated().sum(), precipitation.duplicated().sum(), wildfires.duplicated().sum())

# and nulls

display(air_temp.isnull().sum(), fire_causes.isnull().sum(), negligence.isnull().sum(), precipitation.isnull().sum(), wildfires.isnull().sum())
```

```python
# we need 'state' back to perform the merge

wildfires_df = wildfires_df.rename(columns={'state':'county'})
```

```python
fire_causes_df
```

```python
base_model_df = (wildfires_df.merge(air_temp_df, on=['year', 'month', 'county'], how='left')
                .merge(precipitation_df, on=['year', 'month', 'county'], how='left')
                .merge(fire_causes_df, on=['year', 'month', 'county'], how='left'))
```

```python
fire_causes_df.rename(columns={'burnt_area (ha)':'area per cause'}, inplace=True)
```

```python
fire_causes_df.
```

```python
# now I want to add the causes of fires to the base model, so I need to pivot the table first

pivot_causes = fire_causes_df.pivot_table(index=['year', 'county'], columns='cause', values='area per cause')
pivot_causes.reset_index(inplace = True)
```

```python
pivot_causes.head(1)
```

```python
base_model_df = base_model_df.merge(pivot_causes, on=['year', 'county'], how='left')
base_model_df.head(1)
```

```python
pivot_negligence = negligence_df.pivot_table(index=['year', 'county'], columns='cause', values='burnt_area (ha)')
pivot_negligence.reset_index(inplace = True)
```

```python
pivot_negligence.head(1)
```

```python
base_model_df = base_model_df.merge(pivot_negligence, on=['year', 'county'], how='left')
base_model_df.head(1)
```

```python
base_model_df.rename(columns={'county':'state', 'burnt_area (ha)':'hectares'}, inplace=True)
base_model_df.info()
```

```python
base_model_causes = base_model_df
base_model_df = base_model_df[['year', 'month', 'state', 'fire qty', 'hectares', 'temperature', 'precipitation']]
```
